## 1
Del circuito sabemos que:
$$
I_{1} = \frac{V_{1}-V_{3}}{1/SC} = \frac{V_{3}}{R_{3}} \hspace{0.5cm} (1)\\ 
I_{2} = \frac{V_{1}-V_{3}}{R_{1}} = \frac{V_{3}-V_{2}}{R_{2}} \hspace{0.5cm} (2) 
$$
Por lo que podemos plantear que:
$$
V_{1}-V_{3} = \frac{V_{3}*(1/SC)}{R_{3}} = \frac{R_{1}*(V_{3}-V_{2})}{R_{2}} \hspace{0.5cm} (3)\\ 
\Rightarrow \hspace{0.5cm} V_{3}*(1/SCR_{3} - R_{1}/R_{2}) = -V_{2}*R_{1}/R_{2} \hspace{0.5cm} (4)\\ 
\Rightarrow \hspace{0.5cm} V_{3} = -V_{2}*\frac{R_{1}}{R_{2}*(1/SCR_{3} - R_{1}/R_{2})} \hspace{0.5cm} (5)
$$
Denominamos K para facilitar los calculos
$$
k=\frac{R_{1}}{R_{2}*(1/SCR_{3} - R_{1}/R_{2})}
$$
Asi es que:
$$
V_{3} = -V_{2}*K \hspace{0.5cm} (6)
$$
Ahora tomamos la expresion (6) y la reemplazamos en (1)
$$
\frac{V_{1}+V_{2}*K}{1/SC} = -\frac{V_{2}*K}{R_{3}} \\
\Rightarrow \hspace{0.5cm} V_{1} = - V_{2}*(\frac{1/SC}{R_{3}}+1)*K
$$
Asi es que:
$$
\frac{V_{2}}{V_{3}}=-\frac{1}{ K * ( \frac{1/SC}{R_{3}} + 1)}
$$
Reemplazando K y simplificando llegamos a la **funcion transferencia**
$$
H(S)=\frac{V_{2}}{V_{1}}=\frac{S-R_{2}/R_{1}R_{3}C}{S+1/CR_{3}} \\
|H(j\omega)| = \frac{|j\omega-R_{2}/R_{1}R_{3}C|}{|j\omega+1/CR_{3}|} = \frac{\sqrt[2]{\omega^2-(R_{2}/R_{1}R_{3}C)^2}}{\sqrt[2]{\omega^2+(1/CR_{3})^2}}\\
\angle H(j\omega) = \angle (j\omega-R_{2}/R_{1}R_{3}C) - \angle (j\omega+1/CR_{3}) = \arctan(\frac{\omega}{-R_{2}/R_{1}R_{3}C}) - \arctan(\frac{\omega}{1/CR_{3}})
$$
